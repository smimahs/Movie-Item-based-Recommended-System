```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
movies=pd.read_csv('movies.csv')
movies
```

```python
ratings=pd.read_csv('ratings.csv')
ratings
```

```python
ratings.drop(['timestamp'],axis=1,inplace=True)
```

```python
movies['genres']=movies.genres.str.split('|')
```

```python
movies['title'] = movies.title.str.replace('(\(\d\d\d\d\))', '')
```

```python
movies
```

```python
genredf=movies.copy()
```

```python
for index,row in genredf.iterrows():
    for genre in row['genres']:
        genredf.at[index,genre]=1
genredf=genredf.fillna(0)        
```

```python
user1=ratings.iloc[:3,:]
```

```python
user1
```

```python
profile1=genredf[genredf['movieId'].isin(user1['movieId'])]
```

```python
profile1
```

```python
profile1.drop(['movieId','title','genres'],1,inplace=True)
```

```python
profile1 = profile1.reset_index(drop=True)
```

```python
profile1
```

```python
rating=user1['rating'].values
```

```python
rating1=profile1.transpose().dot(rating)
```

```python
rating1
```

```python
genretable=genredf.set_index('movieId')
```

```python
genretable.drop(['title','genres'],1,inplace=True)
```

```python
genretable
```

```python
recommendationTable1 = ((genretable*rating1).sum(axis=1))/(rating1.sum())
recommendationTable1
```

```python
recommendationTable1.sort_values(ascending=False)
```
